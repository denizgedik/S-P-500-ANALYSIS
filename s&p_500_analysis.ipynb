```python
testing_stocks = testing_stocks[testing_stocks['Adj Close'].notna()]
testing_stocks.isnull().sum()
```

```python
data_combined_head = testing_stocks.groupby(by='Symbol').head(1)
data_combined_tail = testing_stocks.groupby(by='Symbol').tail(1)
data_combined_full = pd.merge(
    data_combined_head, data_combined_tail, on=['Symbol']
)
data_combined_full.head()
```

```python
data_combined = data_combined_full[['Symbol', 'Date_x', 'Open_x', 'Date_y', 'Close_y', 'Industry_x', 'Sector_x']]
data_combined.head(20)
```

```python
data_combined['increase'] = (data_combined['Close_y'] - data_combined['Open_x'])
data_combined['increase_perc'] = ((data_combined['Close_y'] - data_combined['Open_x']) / data_combined['Open_x']) * 100
data_combined.head(20)
```

```python
# Create a loop for each industries using value_counts or something and in each industry, add all the companies from that industries overall increases together, then divide that by number of companies in each industry. This way you will create a median or whatever the fuck for each industries profitability or increase or whatever the fuck.
```

```python
sector_sort = data_combined.sort_values(by='Sector_x')
sector_sort
```

```python
industry_sort = data_combined.sort_values(by=['Sector_x', 'Industry_x'])
industry_sort
```

```python
industry_sort.value_counts('Sector_x')
```

```python
# create charts based on sectors by only allowing companies from certain sectors into their tables.
```

```python
# since there are no more than 2 pieces of price information for each stock, you cannot create a chart for each. But you can show each industry as circles and the bigger industries can be made bigger and biggest companies will also be shaped like a circle
```

```python
# before and after like circles and one will see which ones were big before and which ones are bigger after
```

```python
sector_sort = industry_sort.groupby('Sector_x').mean()
sector_sort
```

```python
industry_new_sort = industry_sort.groupby('Industry_x').mean()
industry_new_sort
```

```python
sector_sort = sector_sort.reset_index()
industry_new_sort = industry_new_sort.reset_index()
```

# DATA ANALYSIS AND VISUALIZATION PHASE

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="whitegrid")

g = sns.PairGrid(sector_sort, y_vars='increase_perc',
                 x_vars='Sector_x',
                 height=6, aspect=3.2)

g.map(sns.pointplot, scale=1.3, errwidth=4, color="xkcd:plum")
sns.despine(fig=g.fig, left=True)
```

```python
sns.set_theme(style="whitegrid")

f, ax = plt.subplots(figsize=(15, 6))

sns.set_color_codes("pastel")
sns.barplot(x="Close_y", y="Sector_x", data=sector_sort,
            label="Close", color="b")

sns.set_color_codes("muted")
sns.barplot(x="Open_x", y="Sector_x", data=sector_sort,
            label="Open", color="b")

handles, labels = ax.get_legend_handles_labels()
ax.legend(handles[::-1], labels[::-1], ncol=2, loc="lower right", frameon=True)
sns.despine(left=True, bottom=True)
```

```python
sns.set_theme(style="white")

sns.relplot(x="Open_x", y="Close_y", hue="Sector_x", size="increase_perc",
            sizes=(40, 400), alpha=.5, palette="muted",
            height=6, data=sector_sort)
```
